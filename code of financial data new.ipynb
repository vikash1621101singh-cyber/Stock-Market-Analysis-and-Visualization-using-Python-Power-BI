```python
threshold = len(df) * 0.4

cols_to_drop = df.columns[df.isnull().sum() > threshold]

print(cols_to_drop)

df.drop(columns=cols_to_drop, inplace=True)
```

```python
print(df.dtypes)
```

```python
num_cols = df.select_dtypes(include=['float64','int64']).columns

for col in num_cols:
    df[col].fillna(df[col].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df[col] = np.clip(df[col], lower, upper)
```

```python
Q1 = df['Gross Profit'].quantile(0.25)
Q3 = df['Gross Profit'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['Gross Profit'] < lower) | (df['Gross Profit'] > upper)]

print("Number of outliers:", len(outliers))
```

```python
df = df[(df['Gross Profit'] >= lower) & (df['Gross Profit'] <= upper)]
```
